**Age and Product Crosstab probabilities**

```python
def age_classification(age):
  AgeRange = np.NaN
  if (age <= 20):
    AgeRange = '20_and_below'
  elif (age >= 21) & (age <= 25):
    AgeRange = '21_to_25'
  elif (age >= 26) & (age <= 30):
    AgeRange = '26_to_30'
  elif (age >= 31) & (age <= 35):
    AgeRange = '31_to_35'
  elif (age >= 36) & (age <= 40):
    AgeRange = '36_to_40'
  elif (age >= 41) & (age <= 45):
    AgeRange = '41_to_45'
  elif (age >= 46) & (age <= 50):
    AgeRange = '46_to_50'
  else:
    AgeRange = '>50'
  return AgeRange

aerofit_df['AgeRange'] = aerofit_df['Age'].apply(age_classification)
aerofit_df
```

```python

age_product_Crosstab  = pd.crosstab(index=aerofit_df.Product, columns=aerofit_df.AgeRange, rownames=['Product'], colnames=['AgeRange'], margins=True, margins_name='Total')
age_product_Crosstab.sort_values(by='Total', axis =1, ascending = False, inplace = True)
age_product_Crosstab.sort_values(by='Total', axis =0, ascending = False, inplace = True)
age_product_Crosstab
```

```python
age_product_Crosstab_conditional_prob = age_product_Crosstab
productlist  = [i for i in list(age_product_Crosstab_conditional_prob.index) if i != 'Total']
agegrouplist = [i for i in list(age_product_Crosstab_conditional_prob.columns) if i != 'Total']

for i in productlist:
  print(f'{10 * "*"} Age given product {i} conditional probabilities {10 * "*"}')
  print('\n')
  for j in agegrouplist:
    prob = age_product_Crosstab_conditional_prob.loc[i, j]/age_product_Crosstab_conditional_prob.loc[i, 'Total']
    print((f'age_{j}_given_{i}' , prob))
  print('\n')

for i in agegrouplist:
  print(f'{10 * "*"} Prodcut given Age group {i} conditional probabilities {10 * "*"}')
  print('\n')
  for j in productlist:
    prob = age_product_Crosstab_conditional_prob.loc[j, i]/age_product_Crosstab_conditional_prob.loc['Total', i]
    print((f'Product_{j}_given_{i}' , prob))
  print('\n')
```

```python

total = age_product_Crosstab.loc['Total', 'Total']
age_product_Crosstab.loc['Total', :'46_to_50'] = age_product_Crosstab.loc['Total', :'46_to_50']/total
age_product_Crosstab.loc['KP281':'KP781', 'Total'] = age_product_Crosstab.loc['KP281':'KP781', 'Total']/total

age_product_Crosstab.rename(index={'Total' : 'Probabilty'}, columns = {'Total' : 'Probabilty'}, inplace = True)
print(f"{10*'*'} Marginal Probabilities {10*'*'}")
age_product_Crosstab
```

**Product and MilesRange Crosstab probabilities**

```python
def miles_classification(miles):
  MilesRange = np.NaN
  if (miles >= 0) & (miles <= 25):
    MilesRange = '0_to_25'
  elif (miles >= 26) & (miles <= 50):
    MilesRange = '26_to_50'
  elif (miles >= 51) & (miles <= 75):
    MilesRange = '51_to_75'
  elif (miles >= 76) & (miles <= 100):
    MilesRange = '76_to_100'
  elif (miles >= 101) & (miles <= 125):
    MilesRange = '101_to_125'
  elif (miles >= 126) & (miles <= 150):
    MilesRange = '126_to_150'
  elif (miles >= 151) & (miles <= 175):
    MilesRange = '151_to_175'
  elif (miles >= 176) & (miles <= 200):
    MilesRange = '176_to_200'
  elif (miles >= 201) & (miles <= 225):
    MilesRange = '201_to_225'
  elif (miles >= 226) & (miles <= 250):
    MilesRange = '226_to_250'
  elif (miles >= 251) & (miles <= 275):
    MilesRange = '251_to_275'
  elif (miles >= 276) & (miles <= 300):
    MilesRange = '276_to_300'
  elif (miles >= 301) & (miles <= 325):
    MilesRange = '301_to_325'
  elif (miles >= 326) & (miles <= 350):
    MilesRange = '326_to_350'
  else:
    MilesRange = '>350'
  return MilesRange

aerofit_df['MilesRange'] = aerofit_df['Miles'].apply(miles_classification)
aerofit_df
```
